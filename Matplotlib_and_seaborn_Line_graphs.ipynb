```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.936, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields In Uganda')
plt.legend(['Apples', 'Oranges']);
```

```python
# Line markers
plt.plot(years, apples, marker = 'o')
plt.plot(years, oranges, marker = 'x')

plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years, apples, marker = 'o')
plt.plot(years, oranges, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields in Uganda')
plt.legend(['Apples', 'Oranges']);
```

# Styling Lines and markers
## .color or c : set the color of the line
## .linestyle or ls : choose between a solid or dashed line
## . linewidth or lw: set the size of markers
## .markersize or ms : set the size of markers
## .markeredgecolor or mec : set the edge color for the markers
## .markeredgewidth or new : set the edge width for the markers
## .markerfacecolor or mfc : set the fill color for the makers
## .alpha : opacity of the plot

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('year')
plt.ylabel('Yields (tons per hectare)')

plt.title('Crop Yields in Uganda')
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, oranges, 'or')
plt.title('Yield of Oranges (tons per hectare)'); # 'or' alone gives dotted lines alone
```
